```python
# Binary categorical columns (Yes/No, or 2 unique values)
label_encode_cols = [
    "self_employed", "family_history", "treatment", "Growing_Stress",
    "Changes_Habits", "Mental_Health_History", "Mood_Swings",
    "Coping_Struggles", "Work_Interest", "Social_Weakness",
    "mental_health_interview"
]
```

```python
# One-hot encoding for low-cardinality categoricals
one_hot_encode_cols = [
    "Gender", "Days_Indoors", "care_options", "country_region", "country_region_code"
]
```

```python
# Numeric columns that may contain NaNs
numeric_cols = [
    "retail_and_recreation_percent_change_from_baseline",
    "grocery_and_pharmacy_percent_change_from_baseline",
    "workplaces_percent_change_from_baseline",
    "residential_percent_change_from_baseline"
]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
# === STEP 2: Label Encode binary columns ===
le = LabelEncoder()
for col in label_encode_cols:
    if data[col].isnull().any():
        data[col].fillna(data[col].mode()[0], inplace=True)
    data[col] = le.fit_transform(data[col])
```

```python
# === STEP 3: One-Hot Encode selected columns ===
data = pd.get_dummies(data, columns=one_hot_encode_cols, drop_first=True)
```

```python
# === STEP 4: Convert other columns to numeric (if object) ===
for col in data.columns:
    if data[col].dtype == 'object':
        try:
            data[col] = pd.to_numeric(data[col], errors='coerce')
        except:
            pass
```

```python
# === STEP 5: Fill missing numeric values ===
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].mean())
```

```python
# === Final Output ===
print(data.dtypes)  # Confirm all are numeric
print(data.head())  # Preview
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Scale your data before clustering
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
# Step 1: Impute all NaN values using mean (you can also use median)
imputer = SimpleImputer(strategy="mean")
data_imputed = imputer.fit_transform(data)
```

```python
# Step 2: Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_imputed)
```

```python

# Step 3: Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(data_scaled)
```

```python
# Step 4: Add cluster labels back to DataFrame
data['cluster'] = labels
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Reduce dimensions to 2D
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)  # use scaled data

# Plot the clusters
plt.figure(figsize=(10, 7))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=labels, cmap='Set1', alpha=0.6)
plt.title("KMeans Clustering Visualization (2D PCA)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
K = range(1, 11)
for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(data_scaled)
    inertias.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertias, marker='o')
plt.title("Elbow Method: Optimal Number of Clusters")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia (Sum of Squared Distances)")
plt.grid(True)
plt.show()
```
