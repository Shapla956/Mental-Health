```python
# Cell 15: KMeans Clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# How many rows are we clustering?
print("Rows:", len(data))

# Try K‑Means on just 1 000 rows so it finishes instantly
sample = data.sample(1_000, random_state=42)

num_cols = sample.select_dtypes(include=['float32','int32','int8'])
num_cols = num_cols.drop(columns=['treatment_flag'], errors='ignore')

scaler = StandardScaler()
X_small = StandardScaler().fit_transform(num_cols.fillna(0))

kmeans_small = KMeans(n_clusters=3, random_state=42)
clusters_small = kmeans_small.fit_predict(X_small)

print("Silhouette on 1 000‑row sample:",
      silhouette_score(X_small, clusters_small))
```

```python
from sklearn.utils import resample

# Full clustering on all rows — but silhouette on 5 000‑row sample
num_cols = data.select_dtypes(include=['float32','int32','int8'])
num_cols = num_cols.drop(columns=['treatment_flag'], errors='ignore')

X_all = StandardScaler().fit_transform(num_cols.fillna(0))

kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
data['cluster'] = kmeans.fit_predict(X_all)

#  ---- lightweight silhouette  ----
SAMPLE_SIZE = 5_000
X_sil, y_sil = resample(
    X_all, data['cluster'],
    n_samples=min(SAMPLE_SIZE, len(data)),
    stratify=data['cluster'],
    random_state=42
)

print("Silhouette on sample:",
      silhouette_score(X_sil, y_sil))
```

```python
# After fitting kmeans above:
print(data['cluster'].value_counts().sort_index())

# Peek at centroids for interpretability
centroids = pd.DataFrame(
    kmeans.cluster_centers_,
    columns=num_cols.columns
)
centroids
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_all)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=data['cluster'], palette='viridis')
plt.title('K-Means Clusters Visualized via PCA')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
# Binary categorical columns (Yes/No, or 2 unique values)
label_encode_cols = [
    "self_employed", "family_history", "treatment", "Growing_Stress",
    "Changes_Habits", "Mental_Health_History", "Mood_Swings",
    "Coping_Struggles", "Work_Interest", "Social_Weakness",
    "mental_health_interview"
]
```

```python
# One-hot encoding for low-cardinality categoricals
one_hot_encode_cols = [
    "Gender", "Days_Indoors", "care_options", "country_region", "country_region_code"
]
```

```python
# Numeric columns that may contain NaNs
numeric_cols = [
    "retail_and_recreation_percent_change_from_baseline",
    "grocery_and_pharmacy_percent_change_from_baseline",
    "workplaces_percent_change_from_baseline",
    "residential_percent_change_from_baseline"
]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
# === STEP 2: Label Encode binary columns ===
le = LabelEncoder()
for col in label_encode_cols:
    if data[col].isnull().any():
        data[col].fillna(data[col].mode()[0], inplace=True)
    data[col] = le.fit_transform(data[col])
```

```python
# === STEP 3: One-Hot Encode selected columns ===
data = pd.get_dummies(data, columns=one_hot_encode_cols, drop_first=True)
```

```python
# === STEP 4: Convert other columns to numeric (if object) ===
for col in data.columns:
    if data[col].dtype == 'object':
        try:
            data[col] = pd.to_numeric(data[col], errors='coerce')
        except:
            pass
```

```python
# === STEP 5: Fill missing numeric values ===
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].mean())
```

```python
# Step 1: Impute all NaN values using mean (you can also use median)
imputer = SimpleImputer(strategy="mean")
data_imputed = imputer.fit_transform(data)
```

```python
# Step 2: Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_imputed)
```

```python

# Step 3: Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(data_scaled)
```

```python
# Step 4: Add cluster labels back to DataFrame
data['cluster'] = labels
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Reduce dimensions to 2D
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)  # use scaled data

# Plot the clusters
plt.figure(figsize=(10, 7))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=labels, cmap='Set1', alpha=0.6)
plt.title("KMeans Clustering Visualization (2D PCA)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
K = range(1, 11)
for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(data_scaled)
    inertias.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertias, marker='o')
plt.title("Elbow Method: Optimal Number of Clusters")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia (Sum of Squared Distances)")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_scores = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42)
    labels_k = km.fit_predict(data_scaled)
    score = silhouette_score(data_scaled, labels_k)
    sil_scores.append(score)

plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), sil_scores, marker='s', color='green')
plt.title("Silhouette Score vs Number of Clusters")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```
